## The Greek Alphabet

Capital letters on the right-hand side are obtained by capitalizing the LaTeX command for the lowercase version. Certain capital letters* are exceptions which have no LaTeX commands. Instead, use their Roman counterparts. For example, to produce a capital chi simply type X. This also applies for the lowercase omicron.

When two versions of the lowercase letter are available, a <code>var</code> prefix can be added to obtain the second version. For example, the two versions of epsilon are <code>\epsilon</code> and <code>\varepsilon</code>.
