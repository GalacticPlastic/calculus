When writing integrals, series using sigma-notation or limits; you often want to specify boundaries. You use the same characters as are used for subscripts and superscripts when specifying boundaries.

When it comes the integrals, the boundaries are placed beside the integral-sign. In some cases you may want to place the boundary below the integral-sign; you can then use the command <code>\limits_{}</code>.

The commands used are <code>\int</code> for integral, <code>\sum</code> for sigma-notation, <code>\lim</code> for limits and <code>\prod</code> products.

Ref: [Math-Linux.com Reference](https://www.math-linux.com/latex-26/faq/latex-faq/article/latex-derivatives-limits-sums-products-and-integrals), [Malin Christersson](http://www.malinc.se/math/latex/basiccodeen.php)

## The Greek Alphabet

Capital letters on the right-hand side are obtained by capitalizing the LaTeX command for the lowercase version. Certain capital letters* are exceptions which have no LaTeX commands. Instead, use their Roman counterparts. For example, to produce a capital chi simply type X. This also applies for the lowercase omicron.

When two versions of the lowercase letter are available, a <code>var</code> prefix can be added to obtain the second version. For example, the two versions of epsilon are <code>\epsilon</code> and <code>\varepsilon</code>.

#### Spaces

In math mode you don't get a space when using space bar, white spaces are simply ignored. If you need a space, there are three commands; <code>\\,</code> for a short space, <code>\\:</code>  for a medium space and <code>\\;</code>  for a long space. If you want to reduce the space between two characters, you use the command <code>\\!;</code>  this command is useful for reducing the space between the integral-sign and the integrand.
